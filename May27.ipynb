```python
# profit/loss
```

```python
purchase_cost = 100
selling_cost = 200

if purchase_cost < selling_cost:
    print("profit")
else:
    print("loss")

profit = selling_cost - purchase_cost
print("seller gained this much profit:", profit)
```

```python
# leap year
```

```python
year = int(input("enter the year:"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("century leap year")
        else:
            print("century year but not leap year")
    else:
        print("leap year")
else:
    print("not leap year")
```

```python
# menu square/cube/+/-
```

```python
print("welcome")
num = int(input("enter a num:"))

print("menu\n1.square\n2.cube\n3.check positive or negative")
ch = int(input("select your choice:"))

if ch == 1:
    print(num * num)
elif ch == 2:
    print(num ** 3)
elif ch == 3:
    if num > 0:
        print("positive")
    else:
        print("negative")
else:
    print("please enter correct no")
```

```python
# string Menu
```

```python
char = input("enter a string:")

print("menu\n1.uppercase\n2.lowercase\n3.replace a particular part of the string")
ch = int(input("please enter your choice:"))

if ch == 1:
    print(char.upper())
elif ch == 2:
    print(char.lower())
elif ch == 3:
    r1 = input("which part do you want to replace:")
    r2 = input("what you want to replace it with:")
    new = char.replace(r1, r2)
    print("result:", new)
```
