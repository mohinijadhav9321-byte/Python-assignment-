```python
def vowel_counter(text):
    vowels = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}
    total = 0

    for ch in text.lower():
        if ch in vowels:
            vowels[ch] += 1
            total += 1

    print("Total Vowels =", total)
    print("Vowel Counts =", vowels)

vowel_counter("Hello Python Programming")
```

```python
def palindrome_check(text):
    rev = []

    for ch in text:
        rev.insert(0, ch)

    rev_str = "".join(rev)

    if text == rev_str:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome_check("madam")
```

```python
def char_frequency(text):
    freq = {}

    for ch in text:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    print(freq)

char_frequency("hello")
```

```python
def classify_words(sentence):
    short_words = []
    medium_words = []
    long_words = []

    for word in sentence.split():
        if len(word) < 5:
            short_words.append(word)
        elif len(word) <= 8:
            medium_words.append(word)
        else:
            long_words.append(word)

    print("Short =", short_words)
    print("Medium =", medium_words)
    print("Long =", long_words)

classify_words("Python is a powerful programming language")
```

```python
def matching_strings(lst):
    result = []

    for word in lst:
        if word.startswith("py") and word.endswith("thon"):
            result.append(word)

    print(result)

matching_strings(["python", "python", "pythan", "java"])
```

```python

import string

def unique_word_extractor(paragraph):

    paragraph = paragraph.lower()


    for p in string.punctuation:
        paragraph = paragraph.replace(p, "")


    words = paragraph.split()


    unique_words = set()
    for word in words:
        unique_words.add(word)


    result = sorted(list(unique_words))


    print(result)



text = "Hello world! Hello Python. Python is fun, and world is big."
unique_word_extractor(text)
```

```python
def even_odd_index_separator(text):
    even_chars = []
    odd_chars = []

    for index, char in enumerate(text):
        if index % 2 == 0:
            even_chars.append(char)
        else:
            odd_chars.append(char)

    print("Even indexed characters:", even_chars)
    print("Odd indexed characters:", odd_chars)


# Example usage
text = "PythonProgramming"
even_odd_index_separator(text)
```

```python
def check_anagram(str1, str2):

    list1 = list(str1.lower())
    list2 = list(str2.lower())


    list1.sort()
    list2.sort()


    if list1 == list2:
        print("The strings are anagrams of each other.")
    else:
        print("The strings are NOT anagrams of each other.")



check_anagram("listen", "silent")
```

```python
def missing_characters(text):

    alphabet_set = set("abcdefghijklmnopqrstuvwxyz")


    text = text.lower()


    for char in text:
        if char in alphabet_set:
            alphabet_set.remove(char)


    print("Missing letters:", sorted(list(alphabet_set)))



missing_characters("hello world")
```

```python
def reverse_sentence(sentence):
    words = []

    # S
    current_word = ""

    for char in sentence:
        if char != " ":
            current_word += char
        else:
            words.append(current_word)
            current_word = ""


    words.append(current_word)

    reversed_words = words[::-1]


    result = " ".join(reversed_words)

    print(result)



reverse_sentence("Hello World")
```
