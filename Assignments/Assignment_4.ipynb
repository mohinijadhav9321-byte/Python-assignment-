```python
def safe_divide(a, b):
    try:
        print("Result =", a / b)

    except ZeroDivisionError:
        print("Cannot divide by zero!")

    except TypeError:
        print("Please enter numbers only!")

    finally:
        print("Operation attempted.")


# Valid division
safe_divide(10, 2)

# Division by zero
safe_divide(10, 0)

# String input
safe_divide("10", 2)
```

```python
class InvalidAgeError(Exception):
    def validate_age(age):
        if age < 0 or age > 120:
            raise InvalidAgeError
        print("Valid Age")
    try:
        validate_age(25)
        validate_age(-5)
        validate_age(150)

    except InvalidAgeError:
        print("Invalid age! Age should be between 0 and 120.")
```

```python
# 3. Write a program that repeatedly asks the user (use a loop with at least 3 simulated inputs in a list, e.g. user_inputs = ['10', 'abc', '0', '5']) to enter a number and divide 100 by it. Handle the following in a single try block using multiple except clauses:
#a) ValueError if the input is not a number.
#b) ZeroDivisionError if the input is zero.
#c) A generic Exception as a catch-all for anything else.
#Print the result if successful, otherwise print which error occurred.
user_inputs = ['10', 'abc', '0', '5']

for i in user_inputs:
    try:
        num = int(i)
        result = 100 / num
        print("Result =", result)

    except ValueError:
        print("ValueError: Input is not a number")

    except ZeroDivisionError:
        print("ZeroDivisionError: Cannot divide by zero")

    except Exception:
        print("Some other error occurred")
```

```python
# 4. Create a 1D NumPy array of integers from 1 to 20. Perform the following:
# a) Reshape it into a 4x5 2D array.
# b) Print the sum of each row and each column.
# c) Extract all elements greater than 10 using boolean indexing.
# d) Replace all even numbers in the array with -1.
import numpy as np

# Create array from 1 to 20
a = np.arange(1, 21)

# a) Reshape into 4x5
b = a.reshape(4, 5)
print(b)

# b) Sum of rows and columns
print("Row Sum:")
print(b.sum(axis=1))

print("Column Sum:")
print(b.sum(axis=0))

# c) Elements greater than 10
print("Elements greater than 10:")
print(a[a > 10])

# d) Replace even numbers with -1
a[a % 2 == 0] = -1
print("After replacing even numbers:")
print(a)
```

```python
# 5. Create two NumPy arrays A and B as shown below and perform:
# A = np.array([[1, 2], [3, 4]]) B = np.array([[5, 6], [7, 8]])
# a) Element-wise addition, subtraction, and multiplication.
# b) Matrix multiplication using np.dot() or the @ operator.
# c) Transpose of A.
# d) The determinant and inverse of A using np.linalg.
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# a) Addition, Subtraction, Multiplication
print("Addition:")
print(A + B)

print("Subtraction:")
print(A - B)

print("Multiplication:")
print(A * B)

# b) Matrix Multiplication
print("Matrix Multiplication:")
print(np.dot(A, B))

# c) Transpose of A
print("Transpose of A:")
print(A.T)

# d) Determinant and Inverse of A
print("Determinant of A:")
print(np.linalg.det(A))

print("Inverse of A:")
print(np.linalg.inv(A))
```

```python
# 6. Generate a NumPy array of 15 random integers between 1 and 100 using np.random.randint() (set a seed of 42 for reproducibility). Then:
# a) Find the mean, median, standard deviation, min, and max.
# b) Sort the array in descending order.
# c) Find the indices of the 3 largest values using np.argsort().
# d) Normalize the array so all values lie between 0 and 1.
import numpy as np

np.random.seed(42)

a = np.random.randint(1, 101, 15)

print("Array =", a)

# a) Mean, Median, Min, Max
print("Mean =", np.mean(a))
print("Median =", np.median(a))
print("Std =", np.std(a))
print("Min =", np.min(a))
print("Max =", np.max(a))

# b) Descending order
print(np.sort(a)[::-1])

# c) Indices of 3 largest values
print(np.argsort(a)[-3:])

# d) Normalize
b = (a - a.min()) / (a.max() - a.min())
print(b)
```

```python
# 9. Create a DataFrame with some missing values as shown below and handle them:
# import numpy as np data = { "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"], "Price": [10, np.nan, 5, 8, np.nan], "Stock": [100, 50, np.nan, 30, 20], }
# a) Check the number of missing values in each column using isnull().sum().
# b) Fill missing Price values with the column mean.
# c) Drop rows where Stock is still missing.
# d) Save the cleaned DataFrame to a CSV file named cleaned_products.csv without the index.
import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

# Missing values
print(df.isnull().sum())

# Fill missing Price values with mean
df["Price"] = df["Price"].fillna(df["Price"].mean())

# Drop rows with missing Stock
df = df.dropna(subset=["Stock"])

# Save to CSV
df.to_csv("cleaned_products.csv", index=False)

print(df)
```
