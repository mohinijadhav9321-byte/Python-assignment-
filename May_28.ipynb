```python
# factorial (while loop)
```

```python
num=5
fact=1
while num>0:
    fact=fact*num
    num-=1
print(fact)
```

```python
# sum of digits (12345)
```

```python
num=12345
count=1
sum=0
while count<=5:
    sum=sum+num%10
    count+=1
    num=num//10
print(sum)
```
