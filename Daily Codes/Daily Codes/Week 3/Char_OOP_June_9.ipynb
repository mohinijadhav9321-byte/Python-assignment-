```python
"""
Create a class called Student
Create 2-3 objects having attributes name,age

Create a method which will add a skill (eg...Python, ML, Java) to the student
where you can view the info of it later(i.e. skills learned by a student).
"""
```

```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        self.skill=[]
    def add_skill(self,skill):
        self.skill.append(skill)
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Skills learned by student: {self.skill}")
```

```python
s1=Student("Shreya",17)
s2=Student("Amit",20)
```

```python
s1.add_skill('Python')
```

```python
s1.display_info()
```

```python
s2.add_skill("Java")
```

```python
s2.display_info()
```

```python
s2.add_skill("ML")
```

```python
s2.display_info()
```

```python
s3=Student("Maya",20)
```

```python
s3.display_info()
```

```python
#Inheritance
```

```python
class Person:
    insti="MITU"
    def __init__(self,age,ht):
        self.age=age
        self.ht=ht
    def show(self):
        print(f"Age -> {self.age}, Height -> {self.ht},Institute -> {self.insti}")
```

```python
class Rahul(Person):
    def __init__(self,age,ht,sal):
        super().__init__(age,ht)
        self.sal=sal
    def show_sal(self):
        print(f"Salary is {self.sal}")
```

```python
a=Rahul (29,5.9,5600)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
# Simple menu driven app with oop
```

```python
class Menuapp:
    def __init__(self):
        self.item_list = []

    def show_menu(self):
        print("\nMenu")
        print("1. Add Items")
        print("2. View Items")
        print("3. Exit")

    def add_items(self):
        item = input("Enter item name: ")
        self.item_list.append(item)

    def view_items(self):
        print("Items:", self.item_list)

    def run(self):
        while True:
            self.show_menu()
            ch = int(input("Enter the choice: "))

            if ch == 1:
                self.add_items()

            elif ch == 2:
                self.view_items()

            elif ch == 3:
                print("Exiting...")
                break

            else:
                print("Invalid choice")
```

```python
#Encapsulation with getter and setter
```

```python
class Person:
    def __init__(self,name,age):
        self.__age=age
        self.__name=name

    #getter
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    #setter
    def set_age(self,age):
        self.__age=age
```

```python
p=Person('Arun',34)
```

```python
p.get_name()
```

```python
p.get_age()
```

```python
p.set_age(32)
```

```python
p.get_age()
```
