```python
l2 = [21, 100, 200, 300, 89, 11]

for i in l2:
    print(i**2, end=" ")
print()

for i in l2:
    if i % 2 == 1:
        print(i, end=" ")
print()
```

```python
l2 = [21, 100, 200, 300, 89, 11]

sq = [i**2 for i in l2]
print(sq)

odd = [i for i in l2 if i % 2 == 1]
print(odd)
```
